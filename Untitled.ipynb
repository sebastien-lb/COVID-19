# COVID Analytics

```python
# Read Data from CSV - deaths
import csv
import matplotlib.pyplot as plt

headers = []
data_deaths = []
with open("csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv", "r") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    first = True
    for row in csv_reader:
        if first:
            headers = row
            first = False
            continue
        data_deaths += [row]
print(headers)
print(data_deaths)
```

## Macro Analytics - Global Death

```python
# date start in position 5
nb_day = len(data_deaths[0][4:])
nb_entries = len(data_deaths)
deaths_by_day = [sum([int(data_deaths[j][4 + i]) for j in range(nb_entries)]) for i in range(nb_day)]
```

### Deaths by day

```python
import datetime
import math

def getTime(date):
    try:
        return datetime.datetime.strptime(date,"%m/%d/%y").date()
    except:
        return datetime.datetime.strptime("0" + date,"%m/%d/%y").date()

dates = [getTime(d) for d in headers[4:]]
plt.plot(dates, deaths_by_day)
```

### Deaths by day - Log

```python
plt.plot(dates, [math.log(x) for x in deaths_by_day])
```

```python
# Derivative
derivative_sequence = [deaths_by_day[i+1]-deaths_by_day[i] for i in range(len(deaths_by_day)-1)]
plt.plot(dates[1:], derivative_sequence)
```

```python
# Increase rapport
increase_rapport = [deaths_by_day[i+1]/deaths_by_day[i] for i in range(len(deaths_by_day)-1)]
plt.plot(dates[1:], increase_rapport)
```
